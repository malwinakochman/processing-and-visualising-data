```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
employees = pd.read_csv("employees.csv")
employees
```

```python
employees_selected = employees.groupby(by="sex")
employees_mean = employees_selected['age'].mean()
employees_mean.plot(kind='bar', title='Age avarage by sex', ylabel='Age', xlabel='Sex')
```

```python
employees_by_sex= employees.groupby(by="sex").sum('salary')
employees_by_sex.hist()
```
